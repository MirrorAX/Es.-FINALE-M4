```python
import pandas as pd


file_finale = "owid-covid-data.csv"

es_finale = pd.read_csv(file_finale)

pd.DataFrame(es_finale)
```

```python
#1
es_finale.shape
es_finale.columns
```

```python
#2 ho trovato questo metodo, ma non mi soddisfaceva, perchÃ¨ ripetitivo e macchinoso
europe = es_finale.loc[es_finale["continent"]=="Europe"]
europe_1 = europe.dropna(subset="total_cases")

africa = es_finale.loc[es_finale["continent"]=="Africa"]
africa_1 = africa.dropna(subset="total_cases")

asia = es_finale.loc[es_finale["continent"]=="Asia"]
asia_1 = asia.dropna(subset="total_cases")

nord_america = es_finale.loc[es_finale["continent"]=="North America"]
nord_america_1 = nord_america.dropna(subset="total_cases")

sud_america = es_finale.loc[es_finale["continent"]=="South America"]
sud_america_1 = sud_america.dropna(subset="total_cases")

oceania = es_finale.loc[es_finale["continent"]=="Oceania"]
oceania_1 = oceania.dropna(subset="total_cases")



casi_europa = max(europe_1["total_cases"])
casi_africa = max(africa_1["total_cases"])
casi_asia = max(asia_1["total_cases"])
casi_nord_america = max(nord_america_1["total_cases"])
casi_sud_america = max(sud_america_1["total_cases"])
casi_oceania = max(oceania_1["total_cases"])

print("i casi totali di Covid in Europa sono",casi_europa)
print("i casi totali di Covid in Africa sono",casi_africa)
print("i casi totali di Covid in Asia sono",casi_asia)
print("i casi totali di Covid nel Nord America sono",casi_nord_america)
print("i casi totali di Covid nel Sud America sono",casi_sud_america)
print("i casi totali di Covid in Oceania sono",casi_oceania)
```

```python
#2 ho voluto provare con un ciclo for
zone = ("Europe", "Africa", "Asia", "North America", "South America", "Oceania")
print("Casi per continente:")

for continente in zone:
    continente_df = es_finale.loc[es_finale["continent"] == continente] 
    casi_totali = max(continente_df.dropna(subset="total_cases")["total_cases"])
    print(f"continente: {continente} casi_totali: {casi_totali}")
```
